```python
import requests
from pprint import pprint

BASE_URL = "https://6kb2p9kgb0.execute-api.eu-west-2.amazonaws.com/production/api/v1/addresses/"

api_key = input("Token: ")

params = {
    "postcode": "E8 3LH"
}

response = requests.get(
    BASE_URL,
    params=params,
    headers= {
        "Authorization" : api_key
    }
)

address_list = response.json()
pprint(address_list)
```
